# **Binning**

```python
dataframe
```

```python
dataframe['Discussion'].value_counts()
```

```python
min_val=dataframe['Discussion'].min()
max_val=dataframe['Discussion'].max()
print(min_val)
print(max_val)
```

```python
bins=np.linspace(min_val,max_val,4)
bins
```

```python
labels=['small','medium','big']
```

```python
dataframe['bins']=pd.cut(dataframe['Discussion'],bins=bins,labels=labels,include_lowest=True)
```

```python
dataframe['bins'].value_counts()
```
