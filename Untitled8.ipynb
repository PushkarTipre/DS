```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print(df['age'].mode())
print(df['embarked'].mode())
print(df['embark_town'].mode())
```

```python
df['age'].fillna(value=24,inplace=True)
df['embarked'].fillna(value='S',inplace=True)
df['embark_town'].fillna(value='Southampton',inplace=True)
```

```python

df=df.drop(['deck'],axis=1,inplace=True)
```

```python
sns.barplot(x='sex',y='survived',data=df)
```

```python
sns.barplot(x='embark_town',y='survived',data=df)
```

```python
sns.catplot(x='sex',y='age',data=df)
```

```python
sns.histplot(data=df['fare'],bins=25,kde=True)
```

```python
sns.countplot(x='pclass',hue='survived',data=df)
```
