```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.isnull().sum()
```

```python
print(df['age'].mode())
print(df['embark_town'].mode())
print(df['embarked'].mode())
```

```python
df['age'].fillna(value=24,inplace=True)
df['embark_town'].fillna(value='Southampton',inplace=True)
df['embarked'].fillna(value='S',inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.drop(['deck'],axis=1)
```

```python
df.isnull().sum()
```

```python
custom_palette=['blue','orange']
sns.barplot(hue='sex',y='survived',data=df,)
```

```python
sns.barplot(hue='embark_town',y='survived',data=df)
```

```python
sns.histplot(data=df['fare'],bins=25,kde=True)
plt.title('Fare')
```

```python
sns.catplot(x='sex',y='age',data=df)
```

```python
sns.countplot(x='pclass',hue='survived',data=df)
```

```python
sns.boxplot(x='sex',y='age',hue='survived',data=df)
plt.legend(title='survived',labels=['Yes','No'])
plt.xlabel('Gender')
plt.ylabel('Age')
plt.title('Comparison')
```
